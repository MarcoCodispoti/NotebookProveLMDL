Implementazione mediante numpy

```python
import numpy as np
from tornado.httputil import parse_request_start_line
from urllib3 import encode_multipart_formdata

class Perceptron:
    def __init__(self, learning_rate=0.01, epochs=10):
        self.lr = learning_rate
        self.epochs = epochs
        self.weights = None
        self.bias = None

    def activation_func(self, x):
        return 1 if x >= 0 else 0

    def fit(self, X, y):
        n_samples, n_features = X.shape
        # Inizializziamo i pesi a zero
        self.weights = np.zeros(n_features)
        self.bias = 0

        for _ in range(self.epochs):
            for idx, x_i in enumerate(X):
                # Calocolo della sommatoria pesata
                linear_output = np.dot(x_i, self.weights) + self.bias
                y_predicted = self.activation_func(linear_output)

                # Regola di aggiornamento (Delta rule)
                update = self.lr * (y[idx] - y_predicted)
                self.weights += update * x_i
                self.bias += update

    def predict(self,X):
        linear_output = np.dot(X, self.weights) + self.bias
        # Applichiamo la funzione di attivazione a tutto il vettore
        return np.array([self.activation_func(i) for i in linear_output])

# Esempio d'uso: Porta Logica AND
X = np.array([[0,0],[0,1], [1,0], [1,1]])
y = np.array([0,0,0,1])

p = Perceptron(learning_rate=0.1, epochs=10)
p.fit(X,y)
print(f"Predizioni AND: {p.predict(X)}")    # Output atteso: [0 0 0 1]
```

Implementazione mediante scikit learn

```python
import numpy as np
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score

# -- 1 Preparazione dei Dati (Porta Logica AND) ---
# Input (X): le quattro combinazioni di input binari
X = np.array([[0,0],[0,1], [1,0], [1,1]])
# Output (y): I risultati attesi per l'operazione AND
y = np.array([0,0,0,1])

# --2 Creazione del modello Scikit Leran --
# Inizializziamo l'oggetto Perceptron
# Parametri principali:
# - eta0: Il learning Rate (tasso di apprendimento), corrisponde a \eta . Default è 1
# - max_iter: Il numero massimo di epoche (passaggi sui dati) Default è 1000
# - tol: Tolleranza per l'arresto anticipato (se l'errore non migliora, si ferma)
# - random_state: Per rendere i risultati riproducibili (stessa inizalizzazione casuale dei pesi)
p_sklearn = Perceptron(eta0=0.1, max_iter=10, random_state=42, verbose=1)

# --- 3. Addestra,emto (Fititng) del modello --
# Il metodo fit() esegue l'algoritmo di apprendimento
print("Inizio addestramento..")
p_sklearn.fit(X,y)
print("Addestramento completato. \n")

# --4. Predizione ---
# Utilizziamo il modello addestrato per fare previsioni sugli stessi dati di input
y_pred = p_sklearn.predict(X)

# -- 5. Valutazione e Output dei Risultati --
print("---Risultati---")
print(f"Input X: \n{X}")
print(f"Target reali y: {y}")
print(f"Predizione y_pred: {y_pred}")

# Calcoliamo l'accuratezza (frazione di predizioni corrette)
accuracy = accuracy_score(y,y_pred)
print(f"Accuratezza: {accuracy * 100}%")

# E' possibile ispezionare i pesi e il bias appresi dal modello:
print(f"\nPesi appresi (w1,w2): {p_sklearn.coef_[0]}")
print(f"Bias appreso (b): {p_sklearn.intercept_[0]}")
```

Implementazione del per con metriche di valutazione

```python
import numpy as np
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score

# Definamo i dati di ingresso per la porta logica AND
X = np.array([[0,0],[0,1], [1,0], [1,1]])
y = np.array([0,0,0,1])

# Inizializziamo il modello con un learnig rate di 0.1
# Il parametro random_state garantisce che i risultati siano riproducibili
modello = Perceptron(eta0=0.1, max_iter=10, random_state=42)

# Addestriamo il percettrone: scikit-learn calcola pesi e bias internamente
modello.fit(X,y)

# Eseguiamo le predizioni sul set di dati iniziale
predizioni = modello.predict(X)

# Eseguiamo le predizioni sul set di dati iniziale
precisione = accuracy_score(y, predizioni)

print(f"Risultati predizioni: {predizioni}")
print(f"Accuratezza finale: {precisione * 100}%")
print(f"\nPesi finale: {modello.coef_} | Bias: {modello.intercept_}")
```
