```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.metrics import silhouette_score

# Creazione di dati sintetici
X, _ = make_blobs(n_samples=50, centers=3, cluster_std=0.60, random_state=42)

# Calcolo del legame (linkage)
# 'ward' minimizza la varianza dei cluster che vengono fusi
Z = linkage(X, method='ward')

plt.figure(figsize=(10, 5))
dendrogram(Z)
plt.title("Dendrogram Gerarchico")
plt.xlabel("Indice del campione")
plt.ylabel("Distanza di Ward")
plt.show()

# Visualizzazione del modello
model = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')

# Fit e predict
labels = model.fit_predict(X)

# Visualizzazione dei risultati
plt.scatter(X[:,0], X[:,1], c=labels, cmap='viridis', edgecolors='k')
plt.title("Risultati del Clusteting Agglomerativo")
plt.title("Risultato Clustering Agglomerativo")
plt.show()
```

```python
score = silhouette_score(X, labels)
print(f"Silhouette score: {score: .2f}")
```

```python
from scipy.spatial.distance import pdist
from scipy.cluster.hierarchy import cophenet

# 3. Calcolo del Cophnetic Correlation Coefficient
# pdist(X) calcola le distanze euclidee originale tra i punti (pairwise distances)
# Cophenet(Z,...) calcola le distanze nel dendrogramma e restituisce il coefficiente
c, coph_dists =cophenet(Z,pdist(X))

print(f"Cophenetic Correlation Coefficient: {c: .4f}")
```
