Implementing decision tre using scikit learn and libraries

```python
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, ConfusionMatrixDisplay

# Dataset di esempio
# caricare il dataset Iris
iris = load_iris()
X = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Modello decision Tree criterio Gini
clf = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=42)
clf.fit(X_train, y_train)

# Predizione e valutazione
y_pred = clf.predict(X_test)
print("Accuracy: ", accuracy_score(y_test, y_pred))

# Visualizzazione dell'albero
plt.figure(figsize=(10, 6))
plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.show()


# Valutare la performance
print("Accuracy: ", accuracy_score(y_test, y_pred))
print("\nClassification Report: \n", classification_report(y_test, y_pred, target_names=iris.target_names))
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=iris.target_names)
disp.plot(cmap='Blues')
plt.title("Matrice di confusione - Decision Tree")
plt.show()
```

Visualizing features ordered by importance

```python
import pandas as pd
import numpy as np

importance = pd.DataFrame({
    'Feature': load_iris().feature_names,
    'Importance': np.round(clf.feature_importances_, 3)
}).sort_values(by='Importance', ascending=False)

print(importance)
```
